## 4. Data Preprocessing & Training data development

#### Creating a function to convert the text in lowercase, remove the extra space, special char, url and links.

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)    
    return text
```

```python
df["text"] = df["text"].apply(wordopt)
```

```python
df.dtypes
```

#### Splitting the dataset into training set and testing set.

#### Convert text to vectors

```python
X_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
XV_train = vectorization.fit_transform(X_train)
XV_test = vectorization.transform(X_test)
```

## 5.Modeling

### a). Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(XV_train,y_train)
```

```python
pred_logreg=logreg.predict(XV_test)
```

```python
logreg.score(XV_test, y_test)
```

```python
print(classification_report(y_test, pred_logreg))
```

```python
cm = confusion_matrix(y_test, pred_logreg)
print(cm)
```

### prediction  using news data

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "REAL NEWS"
    
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt) 
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_logreg= logreg.predict(new_xv_test)
    pred_dectree = dectree.predict(new_xv_test)
    pred_gradboostclsf = gradboostclsf.predict(new_xv_test)
    pred_ranforclsf = ranforclsf.predict(new_xv_test)

    return print("\n\nLogistic Prediction: {} \nDecisionTreeClassifier Prediction: {} \nGradietBoosting Prediction: {}\
    \nRandomForest Classifier Prediction: {}".format(output_label(pred_logreg[0]), 
                                output_label(pred_dectree[0]), 
                                output_label(pred_gradboostclsf[0]), 
                                output_label(pred_ranforclsf[0])))
```

```python
news = str(input(("Enter the news text to predict")))
manual_testing(news)
```
